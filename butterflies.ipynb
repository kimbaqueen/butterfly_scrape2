## How is the site organized? If you want to create a butterfly dataset manually, how would you do it?

Site is organized with a header and multiple view pages. You can view the butterflies library by A-Z, location, and other categories. Once you click on a butterfly, it opens it up in a seperate page view with information that is mostly provided via narrative paragraph style. Manually this would be challenging and time intensive to extract all relevant data. There is a way to save a PDF, but not CSV or excel format, and the data is not in a table format. Not easy to extract. Would have to manually create a table and copy data line by line into it for each species.

```python
def get_butterfly(url):
    """Request and parse a single butterfly profile page, return a dict of data."""


    r = requests.get(url)
    soup = BeautifulSoup(r.text)

    family = soup.find("li", text=re.compile(r'Family:*'))
    size = soup.find("li", text=re.compile(r'Size:*'))
    wing_span = soup.find("li", text=re.compile(r'Wing Span:*'))



    return {
            'name': name,
            'family': peel_data_from_element(family),
            'size': peel_data_from_element(size),
            'wing span': peel_data_from_element(wing_span),
            'url': url,
           }


def peel_data_from_element(element):
  just_text = element.text
  return just_text.split(': ')[1]
```

```python
def write_csv():
    
    to_csv = [
        {'name': soup.find()},
        {'url': soup.find()},
        {'family': soup.find("li", text=re.compile(r'Family:*'))},
        {'size': soup.find("li", text=re.compile(r'Size:*'))},
        {'wing_span': soup.find("li", text=re.compile(r'Wing Span:*'))},
    ]
    
    keys = to_csv[0].keys()

    with open('butterflies.csv', 'w', newline='') as output_file:
        dict_writer = csv.DictWriter(output_file, keys)
        dict_writer.writeheader()
        dict_writer.writerows(to_csv)
    
```
